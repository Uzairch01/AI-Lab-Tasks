```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
# 1. IMDB Dataset (Sentiment analysis of movie reviews)
def imdb_analysis():
    imdb_data = load_dataset("imdb")
    df_imdb = pd.DataFrame(imdb_data['train'])

    # Basic analysis of sentiment distribution
    sentiment_counts = df_imdb['label'].value_counts()

    # Plotting sentiment distribution
    plt.figure(figsize=(6, 4))
    sentiment_counts.plot(kind='bar', color=['blue', 'orange'])
    plt.title("IMDB Sentiment Distribution")
    plt.xlabel("Sentiment (0 = Negative, 1 = Positive)")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
```

```python
imdb_analysis()
```

```python
# 2. AG News Dataset (News articles by genre)
def ag_news_analysis():
    ag_news_data = load_dataset("ag_news")
    df_ag_news = pd.DataFrame(ag_news_data['train'])

    # Basic analysis of news category distribution
    category_counts = df_ag_news['label'].value_counts()

    # Plotting category distribution
    plt.figure(figsize=(6, 4))
    category_counts.plot(kind='bar', color='green')
    plt.title("AG News Category Distribution")
    plt.xlabel("News Category (0 = World, 1 = Sports, etc.)")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
```

```python
ag_news_analysis()
```

```python
# 3. Amazon Polarity Dataset (Amazon reviews categorized by sentiment)
def amazon_polarity_analysis():
    amazon_data = load_dataset("amazon_polarity")
    df_amazon = pd.DataFrame(amazon_data['train'])

    # Basic analysis of sentiment distribution
    sentiment_counts = df_amazon['label'].value_counts()

    # Plotting sentiment distribution
    plt.figure(figsize=(6, 4))
    sentiment_counts.plot(kind='bar', color=['purple', 'red'])
    plt.title("Amazon Polarity Sentiment Distribution")
    plt.xlabel("Sentiment (0 = Negative, 1 = Positive)")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
```

```python
amazon_polarity_analysis()
```

```python
# 4. SQuAD Dataset (Question-answer pairs)
def squad_analysis():
    squad_data = load_dataset("squad")
    df_squad = pd.DataFrame(squad_data['train'])

    # Analysis: Length of context and question
    df_squad['context_length'] = df_squad['context'].apply(len)
    df_squad['question_length'] = df_squad['question'].apply(len)

    # Plotting histogram of context and question length
    plt.figure(figsize=(6, 4))
    df_squad['context_length'].hist(bins=50, color='blue', alpha=0.7, label='Context Length')
    df_squad['question_length'].hist(bins=50, color='red', alpha=0.5, label='Question Length')
    plt.title("SQuAD Context and Question Length Distribution")
    plt.xlabel("Length")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```

```python
squad_analysis()
```

```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Define the function to load and analyze the CIFAR-10 dataset
def cifar10_analysis():
    # Load the CIFAR-10 dataset
    cifar10_data = load_dataset("cifar10")

    # Convert the dataset into a Pandas DataFrame
    df_cifar10 = pd.DataFrame(cifar10_data['train'])

    # Check the first few rows and column names
    print(df_cifar10.head())
    print(df_cifar10.columns)  # Check columns

    # Assuming the labels (image classes) are in a column named 'label'
    class_counts = df_cifar10['label'].value_counts()

    # Plotting the class distribution
    plt.figure(figsize=(8, 5))
    class_counts.plot(kind='bar', color='skyblue')
    plt.title("CIFAR-10 Image Class Distribution")
    plt.xlabel("Image Class")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

# Call the function to analyze and plot the CIFAR-10 dataset
cifar10_analysis()
```
